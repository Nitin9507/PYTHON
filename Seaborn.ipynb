```python
import seaborn as sc
```

```python
import matplotlib.pyplot as plt
```

```python
df = sc.load_dataset("tips")
df
```

```python
sc.histplot(df["tip"], kde=True)
plt.show
```

```python
s= sc.load_dataset("tips")
sc.scatterplot(x="total_bill", y="tip", data=s)
```

```python
s= sc.load_dataset("tips")
sc.scatterplot(x="total_bill", y="tip", hue="sex", size="size", data=s)
plt.title("Relationship between Total Bill and Tip")
plt.show()
```

```python
sc.histplot(df['sex'], color="gray", edgecolor="Blue")
```

```python
fmri= sc.load_dataset("fmri")
fmri
```

```python
sc.lineplot(x="timepoint", y="signal", marker= "*", ms=10, mfc="red", linestyle="--", data=fmri)
```

```python
sc.lineplot(x="timepoint", y="signal", hue="event",marker= "*", ms=10, mfc="red", style="region",linestyle="--", data=fmri)
```

```python
df= sc.load_dataset("titanic")
df
```

```python
sc.barplot(x="age", y="class", color="skyblue", hue="sex", errorbar=None, palette="muted", data=df)
```

```python
p= sc.load_dataset("iris")
p
```
